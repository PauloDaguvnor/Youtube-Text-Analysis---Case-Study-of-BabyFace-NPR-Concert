```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

nltk.download("stopwords")

#Define the Text Pre-processing pipeline
def preprocess(row):
    text = row["text"]
    text = text.lower()

    # Remove emojis using a regular expression
    emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)

    text = emoji_pattern.sub(r'', text)  # Remove emojis

    # Remove links using a regular expression
    link_pattern = re.compile(r"http\S+|www\S+")
    text = link_pattern.sub(r'', text)

    # Define a regular expression to remove punctuation marks and symbols
    tokenizer = RegexpTokenizer(r'\w+')
    words = tokenizer.tokenize(text)

    # Remove stopwords
    stop_words = set(stopwords.words("english")+ ['39'] + ['br'])
    words = [word for word in words if word.lower() not in stop_words]

    return ' '.join(words)

#Apply prepocessing steps to the dataframe
en_df["cleaned_body"] = en_df.apply(preprocess, axis=1)
```

```python
# List of names and their variations
names_and_variations = {
    'Kenny Edmonds': ['Kenny', 'baby face', 'babyface'],
    'Avery Wilson': ['Avery'],
    'Tank': ['tank'],
    'Chante Moore': ['Chante', 'Chanté']
}

# Function to count unique occurrences of names and their variations in a text
def count_unique_name_variations(text, name_variations):
    unique_names = set()
    for name_var_list in name_variations.values():
        for name_var in name_var_list:
            matches = re.findall(r'\b' + re.escape(name_var) + r'\b', text, re.IGNORECASE)
            if matches:
                unique_names.add(next(iter(matches)))  # Add only the first occurrence of the name or variation
    return len(unique_names)

# Initialize a dictionary to store the total counts for each name
total_counts = {name: 0 for name in names_and_variations.keys()}

# Apply the function to calculate the counts for each row and update the total_counts dictionary
for name, variations in names_and_variations.items():
    en_df[f'{name}_count'] = en_df['text'].apply(lambda x: count_unique_name_variations(x, {name: variations}))
    total_counts[name] += en_df[f'{name}_count'].max()

# Calculate the total counts for each name and update the total_counts dictionary
for name in names_and_variations.keys():
    total_counts[name] = name_counts_df[name].sum()

# Create a DataFrame with the total counts for each name
total_counts_df = pd.DataFrame(total_counts.items(), columns=['Name', 'Total_Count'])

# Display the DataFrame with the counts for each name and its variations
# print(en_df)
print("\nTotal Counts:")
print(total_counts_df)
```

# Topic Modelling

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import NMF
from sklearn.pipeline import make_pipeline

from sklearn.decomposition import LatentDirichletAllocation
tfidf_vectorizer = TfidfVectorizer(stop_words=stoplist, ngram_range=(2,3))
lda = LatentDirichletAllocation(n_components=5)
pipe = make_pipeline(tfidf_vectorizer, lda)
pipe.fit(kdf2['cleaned_body'])
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        message = "Topic #%d: " % topic_idx
        message += ", ".join([feature_names[i]
                             for i in topic.argsort()[:-n_top_words - 1:-1]])
        print(message)
    print()
print_top_words(lda, tfidf_vectorizer.get_feature_names_out(), n_top_words=3)
```
